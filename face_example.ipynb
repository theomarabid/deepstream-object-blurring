```python
import cv2
import matplotlib.pyplot as plt
```

```python
# path = '/home/omarabid/Downloads/(Canopy) Thu Oct 13 2022 12h 00m 00s (GMT-04_00).mp4'
# path = '/home/omarabid/quantiphi/cfa/data/01072/01072_dt_stack_10_13.mp4'
path = '/home/omarabid/quantiphi/cfa/data/01573/10_13_upload/01573_outside_dt_window_10_13.mp4'
cap = cv2.VideoCapture(path)
```

```python
_, frame = cap.read()
height, width = frame.shape[:2]
```

```python
resized_frame = frame[400:1000, 860:1920, :]
print(resized_frame.shape, resized_frame.shape[1] / resized_frame.shape[0])

plt.imshow(resized_frame)
```

```python
writer = None
for i in range(500):
    _, frame = cap.read()
    resized_frame = frame[400:1000, 860:1920, :]
    height, width = resized_frame.shape[:2]
    if not writer:
        fourcc = cv2.VideoWriter_fourcc('m', 'p', '4', 'v')
        writer = cv2.VideoWriter(path.replace('.mp4', 'crop.mp4'), fourcc,10,(width,height))
    writer.write(resized_frame)
writer.release()
```
