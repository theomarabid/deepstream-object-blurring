```python
_, frame = cap.read()
height, width = frame.shape[:2]
print(frame.shape)
plt.imshow(frame)
```

```python
resized_frame = frame[100:500, 1100:, :]
print(resized_frame.shape, resized_frame.shape[1] / resized_frame.shape[0])

plt.imshow(resized_frame)
```

```python
writer = None
new_path = path.replace('.mp4', 'crop.mp4')
for i in range(2000):
    _, frame = cap.read()
    resized_frame = frame[100:500, 1100:, :]
    height, width = resized_frame.shape[:2]
    if not writer:
        fourcc = cv2.VideoWriter_fourcc('m', 'p', '4', 'v')
        writer = cv2.VideoWriter(new_path, fourcc,10,(width,height))
    writer.write(resized_frame)
writer.release()
print(f"Wrote to {new_path}")
```
